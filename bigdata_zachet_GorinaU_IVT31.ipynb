#### Выполнить расчет частного коэффициента корреляции между выбранными количественными показателями с исключением влияния других количественных признаков датасета.

```python
def matrix_minor(arr, i, j):
    return np.delete(np.delete(arr,i,axis=0), j, axis=1)
def r_partial(arr, i,j):
    aij=(-1)**(i+j)*np.linalg.det(matrix_minor(np.array(arr), i,j))
    aii=np.linalg.det(matrix_minor(np.array(arr), i,i))
    ajj=np.linalg.det(matrix_minor(np.array(arr), j,j))
    return -aij/math.sqrt(aii*ajj)
def corr_partial(arr):
    res=np.copy(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i!=j:
                res[i,j]=r_partial(arr, i,j)
            else:
                res[i,j]=-r_partial(arr, i,j)
    res=pd.DataFrame(res)
    res.index=arr.index;res.columns=arr.columns
    return res
```

```python
display((corr_partial(pd.DataFrame({'Avg-Avg-К9':data['Avg-Avg-К9'],'Avg-Avg-К11':data['Avg-Avg-К11']}).corr())).style.background_gradient(cmap='magma').format(precision=3))
display(corr_partial(pd.DataFrame({'Avg-Avg-К9':data['Avg-Avg-К9'],'Avg-Avg-К11':data['Avg-Avg-К11']}).corr()).iloc[1,0])
```

#### Выполнить канонический корреляционный анализ, подобрав разбиение количественных показателей на две группы так, чтобы между сформированными по этим группам каноническим переменным наблюдалась максимальная корреляция.

```python
from sklearn.cross_decomposition import CCA
W=pd.DataFrame()
W['Avg-Avg-К9']=data['Avg-Avg-К9']
W['Avg-Avg-К101']=data['Avg-Avg-К101']
W['Avg-Avg-К102']=data['Avg-Avg-К102']
display(W)
P=data.iloc[:,1:4]
display(P)
scaler = preprocessing.StandardScaler().fit(W)
Ws=scaler.transform(W)
display(pd.DataFrame(Ws))
scaler = preprocessing.StandardScaler().fit(P)
Ps=scaler.transform(P)
display(pd.DataFrame(Ps))
```

```python
ca = CCA(n_components=3, scale=False, copy=True)
ca.fit(Ps,Ws)
U, V=ca.transform(Ps,Ws)
print(U)
print(V)
print(ca.x_weights_)
print(ca.y_weights_)
print(U[:,0])
print(np.dot(np.matrix(ca.x_weights_[:,0]), np.transpose(Ps)))
print(V[:,0])
print(np.dot(np.matrix(ca.y_weights_[:,0]), np.transpose(Ws)))
display(np.corrcoef(U[:, 0], V[:, 0]))
display(np.corrcoef(U[:, 1], V[:, 1]))
display(np.corrcoef(U[:, 1], V[:, 0]))
display(np.corrcoef(V[:, 0], V[:, 1]))
D=pd.DataFrame()
D["V1"]=V[:,0]
D["V2"]=V[:,1]
D["V3"]=V[:,2]
D["U1"]=U[:,0]
D["U2"]=U[:,1]
D["U3"]=U[:,2]
display(D)
display(D.corr())
```
