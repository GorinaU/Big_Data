```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print('{}\n'.format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193]+'\n')
```

```python
iris_dataset['target_names']
```

```python
iris_dataset['feature_names']
```

```python
pd.DataFrame(iris_dataset['data'])
```

```python
iris_dataset['target'].shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0
)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
import mglearn

iris_df = pd.DataFrame(X_train, columns=iris_dataset['feature_names'])

grr = pd.plotting.scatter_matrix(iris_df, c=y_train,
                        hist_kwds={'bins': 20}, s=20, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
iris_dataset['target_names'][prediction]
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
print('{:.2f}'.format(np.mean(y_pred == y_test)))
```

```python
print('{:.2f}'.format(knn.score(X_test, y_test)))
```
