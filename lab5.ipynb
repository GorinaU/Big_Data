```python
import matplotlib.pyplot as plt
import math
%matplotlib notebook
```

```python
a = 1
b = 2
c = 3
d = 4
f = 3
x = 2
n = 20
step = 1

func_dot = []

for _ in range(20):
    y = (a*x**4 + b*x**3 + c*x**2 + d*x + f) * (a*x**2 + b*x + c)
    func_dot.append((x, y))
    a += step
    b += step
    c += step
    d += step
    f += step
    x += step

axis_x = [i[0] for i in func_dot]
axis_y = [i[1] for i in func_dot]
plt.plot(axis_x, axis_y, color='green', marker='o', linestyle='dashed', linewidth=2,)
plt.show()
```
